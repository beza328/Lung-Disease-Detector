```python
def pixel_histogram(data_dir, class_name):
    class_path = os.path.join(data_dir, class_name)
    sample_img_path = os.path.join(class_path, os.listdir(class_path)[0])
    img = Image.open(sample_img_path).convert('L')  # Grayscale
    pixels = list(img.getdata())
    
    plt.hist(pixels, bins=50, color='gray')
    plt.title(f'Pixel Intensity Distribution - {class_name}')
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")
    plt.show()

pixel_histogram("C:/Users/BEZA/Desktop/Lung-Disease-Detector/Dataset/train", "Tuberculosis")
pixel_histogram("C:/Users/BEZA/Desktop/Lung-Disease-Detector/Dataset/train", "Bacterial Pneumonia")
pixel_histogram("C:/Users/BEZA/Desktop/Lung-Disease-Detector/Dataset/train", "Viral Pneumonia")
```
